```python
import pandas as pd

# Загрузка таблиц
df1 = pd.read_csv('fd_bp.csv')  # Укажите путь к первой таблице
df2 = pd.read_csv('fd_fd.csv')  # Укажите путь ко второй таблице

# Получение первых столбцов (можно изменить индекс, если нужен другой столбец)
col1 = df1.iloc[:, 0].dropna().astype(str).tolist()
col2 = df2.iloc[:, 0].dropna().astype(str).tolist()

# Проверка на одинаковую длину (если требуется)
if len(col1) != len(col2):
    print(f"Предупреждение: количество элементов различается (Таблица 1: {len(col1)}, Таблица 2: {len(col2)})")

# Преобразование в множества для быстрого сравнения
set1 = set(col1)
set2 = set(col2)

# Находим общие элементы
common_elements = set1.intersection(set2)
num_common = len(common_elements)
total_elements = max(len(col1), len(col2))  # Или min(), в зависимости от логики

# Процент совпадения
percentage = (num_common / total_elements) * 100

# Элементы, которые есть в первой таблице, но отсутствуют во второй
missing_in_table2 = set1 - set2

# Элементы, которые есть во второй таблице, но отсутствуют в первой
missing_in_table1 = set2 - set1

# Вывод результатов
print(f"Общее количество элементов: Таблица 1 = {len(col1)}, Таблица 2 = {len(col2)}")
print(f"Количество совпадающих элементов: {num_common}")
print(f"Процент совпадения: {percentage:.2f}%\n")

print("Элементы, которые есть в Таблице 1, но отсутствуют в Таблице 2:")
print(missing_in_table2 if missing_in_table2 else "Нет таких элементов\n")

print("Элементы, которые есть в Таблице 2, но отсутствуют в Таблице 1:")
print(missing_in_table1 if missing_in_table1 else "Нет таких элементов\n")
```
